First load the data into Pandas 

```python
import pandas as pd, geopandas as gpd
import matplotlib.pyplot as plt

# Set up the categories
dict_dataset = {'category_freq' : [0,1,5,6], 'category_labels' : ['A', 'B', 'C', 'D']}

df = pd.DataFrame(data = dict_dataset)
df.plot.barh(x='category_labels', y='category_freq', alpha=0.5, title='Category Frequency Breakdown')
plt.show()
```
